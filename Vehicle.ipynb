```python
#import random
n=0
def freq(a,b,c,d):
    global n
    if n>1:
        n=0
    if n==0:
        plt.figure(figsize=(c,d))
    if b==0:
        plt.subplot(2,1,n+1)
    elif b==1:
        plt.subplot(1,2,n+1)
    n+=1
    fig=car[a].value_counts().plot(kind='bar', color=np.random.rand(len(car[a]),3))
    #fig=car[a].value_counts().plot(kind='bar', color=(random.uniform(0, 1), random.uniform(0, 1), random.uniform(0, 1)))
    plt.title(a+' histogram')
    fig.set(xlabel = a, ylabel='Frequency of '+a)
    plt.tight_layout()
def price_comp(a,b):
    plt.figure(figsize=(28,8))
    plt.title(a+' vs price')
    sns.lineplot(x=car[car[a]<=np.percentile(car[a],b)][a],y=car['price'],color=np.random.rand(3,))
    plt.show()
def price_spread(a,b,c,d):
    global n
    if n>1:
        n=0
    if n==0:
        plt.figure(figsize=(c,d))
    plt.subplot(1,2,n+1)
    n+=1
    plt.title(a+' price spread')
    sns.boxplot(x=car[a],y=car['price'], palette=(b))
    plt.tight_layout()
def price_mean(a,b,c,d):
    global n
    if n>1:
        n=0
    if n==0:
        plt.figure(figsize=(c,d))
    if b==0:
        plt.subplot(2,1,n+1)
    elif b==1:
        plt.subplot(1,2,n+1)
    n+=1
    fig=(car.groupby([a])['price'].mean().sort_values(ascending = False)).plot.bar(color=np.random.rand(len(car[a]),3))
    plt.title(a+' mean price')
    fig.set(xlabel = a, ylabel='Mean price of '+a)
    plt.tight_layout()
```
