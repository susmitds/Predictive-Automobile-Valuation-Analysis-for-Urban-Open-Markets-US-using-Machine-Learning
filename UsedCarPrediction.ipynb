```python
def price_graph(g,double_col,x_size,y_size,var_list,mod):
    plt.figure(figsize=(x_size,y_size))
    if type(var_list) is list:
        for i in range(len(var_list)):
            if double_col==0:
                plt.subplot(len(var_list),1,i+1)
            elif double_col==1:
                plt.subplot(int(np.ceil(len(var_list)/2)),2,i+1)
            draw(g,var_list[i],mod,i)
    else:
        draw(g,var_list,mod,0)
def draw(g,var,mod,i):
    z=(mod[i] if len(mod)>=i else mod[0]) if (type(mod) is list) else mod
    if g=='hist':
        plt.title(var+' histogram')
        (car[var].value_counts().plot(kind='bar', color=np.random.rand(len(car[var]),3))).set(xlabel = var, ylabel='Frequency of '+var)
    elif g=='line':
        plt.title(var+' vs price')
        sns.lineplot(x=car[car[var]<=np.percentile(car[var],z)][var],y=car['price'],color=np.random.rand(3,))
    elif g=='box':
        plt.title(var+' price spread')
        sns.boxplot(x=car[var],y=car['price'], palette=(z))
    elif g=='bar':
        plt.title(var+' mean price')
        ((car.groupby([var])['price'].mean().sort_values(ascending = False)).plot.bar(color=np.random.rand(len(car[var]),3))).set(xlabel = var, ylabel='Mean price of '+var)
    plt.tight_layout()
```

```python
price_graph('hist',0,18,24,['manufacturer','year','paint_color'],0)
price_graph('hist',1,18,24,['condition','cylinder_number','fuel_type','status','transmission','drive_type','size','body_type'],0)
```
