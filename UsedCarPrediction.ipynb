```python
def freq(vlist,b,c,d):
    plt.figure(figsize=(c,d))
    if type(vlist) is list:
        x=len(vlist)
        for i in range(x):
            if b==0:
                plt.subplot(x,1,i+1)
            elif b==1:
                plt.subplot(int(np.ceil(x/2)),2,i+1)
            fig=car[vlist[i]].value_counts().plot(kind='bar', color=np.random.rand(len(car[vlist[i]]),3))
            #fig=car[a].value_counts().plot(kind='bar', color=(random.uniform(0, 1), random.uniform(0, 1), random.uniform(0, 1)))
            plt.title(vlist[i]+' histogram')
            fig.set(xlabel = vlist[i], ylabel='Frequency of '+vlist[i])
    else:
        fig=car[vlist].value_counts().plot(kind='bar', color=np.random.rand(len(car[vlist]),3))
        #fig=car[a].value_counts().plot(kind='bar', color=(random.uniform(0, 1), random.uniform(0, 1), random.uniform(0, 1)))
        plt.title(vlist+' histogram')
        fig.set(xlabel = vlist, ylabel='Frequency of '+vlist)
    plt.tight_layout()
def price_comp(vlist,b,c,d,percentile):
    plt.figure(figsize=(c,d))
    if type(vlist) is list:
        x=len(vlist)
        for i in range(x):
            if b==0:
                plt.subplot(x,1,i+1)
            elif b==1:
                plt.subplot(int(np.ceil(x/2)),2,i+1)
            sns.lineplot(x=car[car[vlist[i]]<=np.percentile(car[vlist[i]],percentile[i])][vlist[i]],y=car['price'],color=np.random.rand(3,))
            plt.title(vlist[i]+' histogram')
    else:
        plt.title(vlist+' vs price')
        sns.lineplot(x=car[car[vlist]<=np.percentile(car[vlist],percentile)][vlist],y=car['price'],color=np.random.rand(3,))
    plt.show()
def price_spread(vlist,b,c,d,color):
    plt.figure(figsize=(c,d))
    if type(vlist) is list:
        x=len(vlist)
        for i in range(x):
            if b==0:
                plt.subplot(x,1,i+1)
            elif b==1:
                plt.subplot(int(np.ceil(x/2)),2,i+1)
            plt.title(vlist[i]+' price spread')
            sns.boxplot(x=car[vlist[i]],y=car['price'], palette=(color[i]))
    else:
        plt.title(vlist+' price spread')
        sns.boxplot(x=car[vlist],y=car['price'], palette=(color))
    plt.tight_layout()
def price_mean(vlist,b,c,d):
    plt.figure(figsize=(c,d))
    if type(vlist) is list:
        x=len(vlist)
        for i in range(x):
            if b==0:
                plt.subplot(x,1,i+1)
            elif b==1:
                plt.subplot(int(np.ceil(x/2)),2,i+1)
        fig=(car.groupby([vlist[i]])['price'].mean().sort_values(ascending = False)).plot.bar(color=np.random.rand(len(car[vlist[i]]),3))
        plt.title(vlist[i]+' mean price')
        fig.set(xlabel = vlist[i], ylabel='Mean price of '+vlist[i])
    else:
        fig=(car.groupby([vlist])['price'].mean().sort_values(ascending = False)).plot.bar(color=np.random.rand(len(car[vlist]),3))
        plt.title(vlist+' mean price')
        fig.set(xlabel = vlist, ylabel='Mean price of '+vlist)
    plt.tight_layout()
```
